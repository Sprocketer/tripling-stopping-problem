```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
max_tries = 10000

random_tries = 1000
```

```python
rewards = {}
for i in range(max_tries):
    tries = i + 1
    failed = False
    earned_list = []
    for _ in range(random_tries):
        earned = 1
        for _ in range(tries):
            generated = random.randint(1, 100)
            if generated >= 51:
                earned *= 3
            else:
                failed = True
                earned = 0
            if failed:
                break
        earned_list.append(earned)
    rewards[tries] = sum(earned_list) / random_tries
print(rewards)
```

# Plotting Average Reward vs. Number of Tries

```python
x = list(rewards.keys())
y = list(rewards.values())
plt.plot(x, y)
plt.triplot(x, y)

window_size = 100
smoothed_y = np.convolve(y, np.ones(window_size)/window_size, mode='valid')# Smooth the y values using moving average
plt.plot(x[window_size-1:], smoothed_y) # Plot the smoothed line

window_size = 1000
smoothed_y = np.convolve(y, np.ones(window_size)/window_size, mode='valid')# Smooth the y values using moving average
plt.plot(x[window_size-1:], smoothed_y) # Plot the smoothed line

plt.xlabel('Number of tries')
plt.ylabel('Average reward')
plt.show()
```

```python
x = list(rewards.keys())
y = [np.median(arr) for arr in rewards.values()]
plt.plot(x, y)
plt.triplot(x, y)

window_size = 100
smoothed_y = np.convolve(y, np.ones(window_size)/window_size, mode='valid')# Smooth the y values using moving average
plt.plot(x[window_size-1:], smoothed_y) # Plot the smoothed line

window_size = 1000
smoothed_y = np.convolve(y, np.ones(window_size)/window_size, mode='valid')# Smooth the y values using moving average
plt.plot(x[window_size-1:], smoothed_y) # Plot the smoothed line

plt.xlabel('Number of tries')
plt.ylabel('Average reward')
plt.show()
```
